```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("files_relevant/all_matches.tsv", sep='\t')
df
```

```python
plt.rcParams['xtick.labelsize'] = 8

df_innings_1 = df.loc[df["innings"] == 1]
df_innings_2 = df.loc[df["innings"] == 2]

plt.figure(figsize=(15,10))

plt.subplot(211)
plt.xticks(rotation=90, ticks = [i for i in range(0, 110, 2)])
plt.title("1st Innings")
plt.hist(df_innings_1["runs"], bins = [x for x in range(110)], color='r', alpha = 0.5)

plt.subplot(212)
plt.xticks(rotation=90, ticks = [i for i in range(0, 110, 2)])
plt.title("2nd Innings")
plt.hist(df_innings_2["runs"], bins = [x for x in range(110)], color='b', alpha = 0.5)

plt.show()
```

```python
unique_venue = df.venue.unique()
count = []

for venue in unique_venue:
    count.append(df.loc[df["venue"]==venue, "venue"].count())

unique_venue = sorted(list(zip(count, unique_venue)), reverse = True)
count = list(zip(*unique_venue))[0]
unique_venue = list(zip(*unique_venue))[1]
    
plt.figure(figsize=(15, 7))
    
plt.xticks(rotation=90)
plt.yticks(ticks=[x for x in range(0, 160, 5)])
plt.title("No. of matches per venue")
plt.bar(x = unique_venue, height = count, color='r', alpha=0.7)
plt.show()
```

```python
mean = []
std = []

for venue in unique_venue:
    mean.append(df.loc[df["venue"]==venue, "runs"].mean())
    std.append(df.loc[df["venue"]==venue, "runs"].std())


plt.figure(figsize=(15, 5))

plt.subplot(121)
plt.xticks(rotation=90)
plt.title("mean")
plt.bar(x=unique_venue, height = mean, color='g', alpha=0.7)

plt.subplot(122)
plt.xticks(rotation=90)
plt.title('std')
plt.bar(x=unique_venue, height = std, color='black', alpha=0.5)

plt.show()
```
