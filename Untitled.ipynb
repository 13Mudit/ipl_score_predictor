```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("files_relevant/all_matches.tsv", sep='\t')
df
```

```python
plt.rcParams['xtick.labelsize'] = 8

df_innings_1 = df.loc[df["innings"] == 1]
df_innings_2 = df.loc[df["innings"] == 2]

plt.figure(figsize=(15,10))

plt.subplot(211)
plt.xticks(rotation=90, ticks = [i for i in range(0, 110, 2)])
plt.title("1st Innings")
plt.hist(df_innings_1["runs"], bins = [x for x in range(110)], color='r', alpha = 0.5)

plt.subplot(212)
plt.xticks(rotation=90, ticks = [i for i in range(0, 110, 2)])
plt.title("2nd Innings")
plt.hist(df_innings_2["runs"], bins = [x for x in range(110)], color='b', alpha = 0.5)

plt.show()
```
